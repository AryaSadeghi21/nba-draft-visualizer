```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load data
proj = pd.read_csv("predictions/linear_pred_2025.csv")
hist = pd.read_csv("cleaned_datasets/all_data.csv")
```

```python
# Compute player-level FPpG std

player_fp_std = hist.groupby("NAME")["FPpG"].std(ddof=1)
pos_fp_std = hist.groupby("POS")["FPpG"].std(ddof=1)
global_fp_std = hist["FPpG"].std(ddof=1)

def choose_sigma(row):
    s_player = player_fp_std.get(row["NAME"], np.nan)
    s_pos = pos_fp_std.get(row["POS"], np.nan)
    mu = row["FPpG_pred"]
    # fallback: 15% of mean or global
    for s in [s_player, s_pos, 0.15 * mu, global_fp_std]:
        if pd.notna(s) and s > 0:
            return s
    return global_fp_std

proj["FPpG_sigma"] = proj.apply(choose_sigma, axis=1)
```

```python
# Define simulation function

rng = np.random.default_rng(42)

def simulate_player(mu_fppg, gp_pred, sigma_fppg, N=10000):
    p = np.clip(gp_pred/82, 0, 1)
    gp_samples = rng.binomial(82, p, size=N)
    fppg_samples = rng.normal(mu_fppg, sigma_fppg, size=N)
    fppg_samples = np.clip(fppg_samples, 0, None)
    totals = gp_samples * fppg_samples
    return {
        "mean": totals.mean(),
        "std": totals.std(ddof=1),
        "p10": np.percentile(totals, 10),
        "p50": np.percentile(totals, 50),
        "p90": np.percentile(totals, 90),
        "samples": totals  # keep if you want raw distribution
    }
```

```python
# Run simulation for all players

results = []
for _, row in proj.iterrows():
    sim = simulate_player(row["FPpG_pred"], row["GP_pred"], row["FPpG_sigma"])
    results.append({
        "NAME": row["NAME"],
        "POS": row["POS"],
        "FP_total_pred": row["FP_total_pred"],
        "MC_mean": sim["mean"],
        "MC_std": sim["std"],
        "MC_p10": sim["p10"],
        "MC_p50": sim["p50"],
        "MC_p90": sim["p90"]
    })

mc_df = pd.DataFrame(results)
mc_df.to_csv("projections/monte_carlo_results.csv", index=False)
mc_df.head()
```

```python
# Example: plot distribution for one player
player = "NIKOLA JOKIC"
sim = simulate_player(
    proj.loc[proj["NAME"] == player, "FPpG_pred"].values[0],
    proj.loc[proj["NAME"] == player, "GP_pred"].values[0],
    proj.loc[proj["NAME"] == player, "FPpG_sigma"].values[0],
    N=10000
)
plt.hist(sim["samples"], bins=50, alpha=0.7)
plt.axvline(sim["mean"], color="red", linestyle="--", label="mean")
plt.title(player + " simulated totals")
plt.legend()
plt.show()
```
